```python

import pandas as pd


data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
    'Product': ['Product A', 'Product B', 'Product A', 'Product C', 'Product B'],
    'Category': ['A', 'B', 'A', 'C', 'B'],
    'Sales': [120, 250, 100, 180, 300]
}


df = pd.DataFrame(data)


df.to_csv('sales_data.csv', index=False)

print("Sample CSV file 'sales_data.csv' created.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('sales_data.csv')


print("Data Preview:")
print(df.head())


print("\nData Info:")
print(df.info())


print("\nMissing Values:")
print(df.isnull().sum())


print("\nSummary Stats:")
print(df.describe())


product_sales = df.groupby('Product')['Sales'].sum()
print("\nTotal Sales by Product:")
print(product_sales)


product_sales.plot(kind='bar', title='Total Sales by Product', color='orange')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.grid(True)
plt.tight_layout()
plt.show()


category_avg = df.groupby('Category')['Sales'].mean()
print("\nAverage Sales by Category:")
print(category_avg)


category_avg.plot(kind='pie', autopct='%1.1f%%', title='Average Sales by Category')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
